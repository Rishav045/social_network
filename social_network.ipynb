```python
import networkx as net
import matplotlib.pyplot as plot
```

```python
g= net.Graph()

g= net.read_pajek('/content/network_blood.net')
net.draw(g)
```

```python
def find_donar_or_blood_bank(reciever,donar):
  list2=[n for n in net.all_neighbors(g,"user0")]
  list2.append(reciever)
  list2.append(donar)
  H = g.subgraph(list2)
  list1=net.shortest_path(H,reciever,donar,weight='weight')
  return list1[1]
```

```python
find_donar_or_blood_bank("user23","user59")
```
